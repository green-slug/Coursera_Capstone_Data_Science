```python
import pandas as pd
import numpy as np
```

### Read Wikipedia page into pandas dataframe

```python
link = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html_table = pd.read_html(link)
df_toronto = html_table[0]
```

### Clean dataframe

```python
# only process cells that have an assigned borough. Ignore cells with a borough that is Not assigned
df_toronto = df_toronto[df_toronto.Borough != "Not assigned"]

# if a cell has a borough but a Not assigned neighbourhood, then the neighbourhood will be the same as borough
df_toronto.Neighbourhood = np.where(df_toronto.Neighbourhood == "Not assigned", df_toronto.Borough, df_toronto.Neighbourhood)

# remove nulls if any
df_toronto = df_toronto[df_toronto['Postal Code'].notnull()]
df_toronto = df_toronto[df_toronto.Borough.notnull()]
df_toronto = df_toronto[df_toronto.Neighbourhood.notnull()]

# reset index
df_toronto.reset_index(drop = True, inplace = True)

# check final result
df_toronto
```
